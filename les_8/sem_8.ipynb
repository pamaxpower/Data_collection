- Загрузите датасет googleplaystore.csv с помощью pandas.

```python
import pandas as pd
```

```python
df = pd.read_csv('./googleplaystore.csv')
```

```python
df.head(5)
```

посмотрим кол-во записей и тип ланных каждого столбца

```python
df.info()
```

общее количество пустых значений в каждом столбце

```python
df.isnull().sum()
```

```python
# вычислим медиану значений в столбце рейтинг
mean_rating = df['Rating'].mean()
mean_rating
```

```python
# заполняем пустые ячейки медианным значением
df['Rating'] = df['Rating'].fillna(mean_rating)
```

```python
df.info()
```

следующим этапом будем приводить данные из столбца 'Installs' в числовой вид

```python
# для этого удалим ',' и '+' и приведем к типу int
df['Installs'] = df['Installs'].str.replace(',','').str.replace('+','').astype(int)
```

```python
df.head(5)
```

```python
df.info()
```

проверим датасет на дубликаты

```python
duplicates = df.duplicated()
duplicates_df = df[duplicates]
duplicates_df           # получим 483 дубликатов
```

удалим все дубликаты

```python
df = df.drop_duplicates()
# и снова проверим их наличие
duplicates = df.duplicated()
duplicates_df = df[duplicates]
duplicates_df       # теперь их нету
```

Задание 3

- Определите выбросы в датасете.
- Примените технику (например, удаление выбросов,
винсоризация) для обработки выбросов.
- Выполните стандартизацию данных по выбранным
переменным.


```python
from scipy import stats
from scipy.stats.mstats import winsorize
from sklearn.preprocessing import StandardScaler

df2 = pd.read_csv('./cleaned_apps.csv')

# отклонение от медианного значения
z_score = stats.zscore(df2['Installs'])
z_score
```

```python
treshold = 3
outliers = df2['Installs'][z_score > treshold]
df2.loc[z_score > treshold, 'Installs'] = df2['Installs'].median()
```

```python
plt.hist(df2['Installs'])
plt.xlabel('Installs')
plt.ylabel('Count')
plt.show()
```

```python
z_score = stats.zscore(df2['Reviews'])
treshold = 3
outliers = df2['Reviews'][z_score > treshold]
df2.loc[z_score > treshold, 'Reviews'] = df2['Reviews'].median()

plt.hist(df2['Reviews'])
plt.xlabel('Reviews')
plt.ylabel('Count')
plt.show()
```

```python
df2.info()
```

```python
reviews_df = df2.loc[df2['Reviews'] > 1000000]
plt.hist(reviews_df['Reviews'])
plt.xlabel('Reviews')
plt.ylabel('Count')
plt.show()
```

```python
plt.hist(winsorize(df2['Rating'], limits=[0.1, 0.1]))
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df3 = pd.read_csv('./googleplaystore.csv')
```

```python
le = LabelEncoder()
df3['TypeEncoder'] = le.fit_transform(df3['Type'])
```

```python
df3['Type'].value_counts()
```

применяем one-hot encoding для возрастного рейтинга приложений

```python
df3 = pd.get_dummies(df3, columns=['Content Rating'], prefix='Content Rating', drop_first=True)
df3
```

посмотрим медианное значение рейтинга приложений в различных категориях, но для подростков

```python
pivot_table = df3.pivot_table(index='Category', columns=['Content Rating_Teen'], values='Rating', aggfunc='mean')
pivot_table
```
