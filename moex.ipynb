```python
import requests
import json
import pandas as pd
import datetime as dt
```

https://iss.moex.com/iss/reference/

```python
url = 'https://moex.com/iss/engines/stock/markets/shares/securites/IMOEX'
# параметры, которые мы можем узнать из документации
symbol = 'MOEX'      # маркировка криптовалюты
interval = '1h'         # интервал графика
# временные интервал от и до
start = str(int(dt.datetime(2023,1,1).timestamp()*1000))
end = str(int(dt.datetime(2023,12,1).timestamp()*1000))
```

```python
# формируем параметры
params = {'symbol': symbol, 
          'interval': interval, 
          'startTime': start, 
          'endTime': end}
```

```python
# делаем get-запрос
response = requests.get(url, params= params)
```

```python
# код статуса
response.status_code
```

```python
# выведем ответ
response.text
```

```python
json_res = json.loads(response.text)
json_res
```

```python
# создать датафрейм из данных
df = pd.DataFrame(json_res)

df.columns = ['datetime', 'open', 'high', 'low', 'close', 'volume','close_time', 'qav', 'num_trades','taker_base_vol', 'taker_quote_vol', 'ignore']
```

```python
# первые 5 значений
df.head()
```

```python
# информация
df.info()
```

```python
df.index = [dt.datetime.fromtimestamp(x/1000.0) for x in df.datetime]
```
