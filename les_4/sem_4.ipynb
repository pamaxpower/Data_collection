```python
from lxml import html
import requests
from pprint import pprint
```

```python
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36'}

url='https://www.ebay.com/'
```

```python
response = requests.get(url+'/b/Fishing-Equipment-Supplies/1492/bn_1851047?_trkparms=parentrq%3A3db32bc618f0adbbe9673e13ffff82c8%7Cpageci%3A3a701745-092c-11ef-ac85-eeffd1c1b05c%7Cc%3A4%7Ciid%3A1%7Cli%3A8874', headers=header)
response.status_code
```

```python
dom = html.fromstring(response.text)
dom
```

```python
# найдем список всех наименований товаров на странице
names = dom.xpath('//h3[@class="s-item__title"]/text()')
```

Теперь "достанем" ссылки на эти товары

```python
ref = dom.xpath('//div[contains(@class, "s-item__info")]/a/@href')
```

Также можно найти эти данные, пользуясь тегом по поиску имени <br>
Если внимательно посмотреть тег h3 (содержащий название товара) является дочерним элементом тега a (где как раз и есть ссылка)

```python
links = dom.xpath('//h3[@class="s-item__title"]/../@href')
```

```python
ref==links  # получили одинаковые результаты
```

```python
# посмотрим цену
prices = dom.xpath('//span[@class="s-item__price"]//text()')
```

```python
add_info = dom.xpath('//span[@class="NEGATIVE"]//text()')
```

```python
items_list = []
items = dom.xpath('//ul[@class="b-list__items_nofooter"]/li')
print(items)
for item in items:
    item_info = {}

    name = item.xpath('.//h3[@class="s-item__title"]/text()')
    link = item.xpath('.//h3[@class="s-item__title"]/../@href')
    price = item.xpath('.//span[@class="s-item__price"]//text()')
    add_info = item.xpath('.//span[@class="NEGATIVE"]//text()')

    item_info['name'] = name
    item_info['link'] = link
    item_info['price'] = price
    item_info['add_info'] = add_info
    items_list.append(item_info)
    
pprint(items_list)
```
